```python
# dimension is 5
from cmath import sqrt
from math import dist

point1 = pd.Series([1,2,3,4,5])
point2 = pd.Series([5,4,3,2,1])

def distance(x, y) -> float:
    distance = 0
    size = x.size
    squaredResult = 0

    for i in range(0, size):
        squaredResult += (y[i] - x[i]) * (y[i] - x[i])
    
    return sqrt(squaredResult)

print(distance(point1, point2))
print(dist(point1,point2))
```

```python
from sklearn import preprocessing 
```

```python
frame = pd.read_csv('C:\\Users\\averu\\github-local\\cs\\python\\iris.csv')
# frame.head(5)

feature1 = frame['sepal_length_cm'].array.reshape(-1,1)

# feature1

# Создание шкалировщика
MinMax_scale = preprocessing.MinMaxScaler(feature_range=(0,1))

# Шкалирование признака
scaled_feature1 = MinMax_scale.fit_transform(feature1)

feature1
```

```python
print("Mean:", round(scaled_feature1.mean()))
print("STD:", scaled_feature1.std())
```

```python
feature2 = frame['sepal_width_cm'].array.reshape(-1,1)

# Создание шкалировщик
scaler = preprocessing.StandardScaler()

# Шкалирование признака
scaled_feature2 = scaler.fit_transform(feature2)

feature2
```

```python
print("Mean:", round(scaled_feature2.mean()))
print("STD:", scaled_feature2.std())
```
