```python
import numpy 
from numpy import random
array = []
for i in range(1,6):
    array.append(random.rand())
# (random.rand() for j in range(1,6))

mean = numpy.mean(array)
median = numpy.median(array)

print('mean --->', mean,'\n' , 'median --->', median)

for element in array:
    print(element)

import matplotlib.pyplot as plt

entry = []
for val in array:
    entry.append(array.count(val))

print(entry)

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = array, y = entry)

plt.show()
```

```python
import math as m
# import sys
# oldStdout = sys.stdout
# sys.stdout = sys.stderr

print("Input your values")
val_1 = int(input())
val_2 = int(input())

print("Input binary operation, which would you like.")
print("Available binary operations are: '+', '-', '*', '/', exp(+), sin(+), cos(-), pow(val_1, val_2).")
operator = str(input())

if (operator == "+"):
    print(val_1, "+", val_2, "=", val_1 + val_2)
elif (operator == "-"):
    print(val_1, "-", val_2, "=", val_1 - val_2)
elif (operator == "*"): 
    print(val_1, "*", val_2, "=", val_1 * val_2)
elif (operator == "/"): 
    print(val_1, "/", val_2, "=", val_1 / val_2)
elif (operator == "exp(+)"): 
    print(val_1, "exp(+)", val_2, "=", m.exp(val_1 + val_2))
elif (operator == "sin(+)"): 
    print(val_1, "sin(+)", val_2, "=", m.sin(val_1 + val_2))
elif (operator == "cos(-)"):
    print(val_1, "cos(-)", val_2, "=", m.cos(val_1 - val_2))
elif (operator == "pow(val_1, val_2)"):
    print(val_1, "pow(val_1, val_2)", val_2, "=", m.pow(val_1,val_2))
else:
    print("You input isn't correct, please try again")  
```
