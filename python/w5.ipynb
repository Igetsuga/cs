```python
import math
from decimal import *
import random
```

```python
class Trigonometry:

    def convert_to_radians(x):
        return x * pi / 180

    def sin(x):
        n = 50
        s = 0
        for i in range(n):
            # Ряд Тейлора для ф-ии синуса
            s1 = ((-1)**(i) * x**(2 * i + 1)) / factorial(2 * i + 1)
            s += s1
        return s

    def cos(x):
        n = 50
        s = 0
        for i in range(n):
            # Ряд Тейлора для ф-ии косинуса
            s1 = ((-1)**(i) * x**(2 * i)) / factorial(2 * i)
            s += s1
        return s

    def tan(x):
        return math.tan(x)

    def atan(x):
        n = 100
        s = 0
        for i in range(n):
            s1 = ((-1)**(i) * x**(2 * i + 1)) / (2 * i + 1)
            s += s1
        return s
    
    def asin(x):
        return math.asin(x)
# other variant via Taylor Series
#         for i in range(n):
#             s1 = factorial(factorial(2*n + 1)) / factorial(factorial(2 * n)) * x**(2 * n + 1)) / (2 * n + 1)**2
#             print(s1)
#             s += s1
#         return s

    def acos(x):
        return pi/2 - math.asin(x)



myVal = random.random()
print(myVal)

myVal_rad = Trigonometry.convert_to_radians(myVal)
print(myVal_rad)
print()

print(Trigonometry.cos(myVal_rad))
print(Trigonometry.sin(myVal_rad))
print()

print(Trigonometry.tan(myVal_rad))
print(Trigonometry.atan(myVal))

```

```python
class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
        
    def print_tree(self):
        if (self.left):
            self.left.print_tree()
        if ( self.right ):
            self.right.print_tree()
        print(self.data)
        

    def insert(self, data):
        if ( data < self.data):
            if ( self.left is None ):
                self.left = Tree(data)
            else:
                self.left.insert(data)
        elif ( data > self.data ):
            if ( self.right is None ):
                self.right = Tree(data)
            else:
                self.right.insert(data)
    

tree = Tree(27)
tree.insert(14)
tree.insert(35)
tree.insert(10)
tree.insert(19)
tree.insert(31)
tree.insert(42)
tree.print_tree()
```

```python
import pandas as pd
from pandas import DataFrame,Series
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import numpy as np
import seaborn as sns
```

```python
# Petrol Dataset
df = pd.read_csv("https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv")
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

```python
from sklearn.tree import DecisionTreeRegressor
X = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
print(X)
print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = DecisionTreeRegressor(criterion='mse')
regressor.fit(X_train, y_train)
```

```python
from sklearn import tree
tree.plot_tree(regressor)
```

```python
y_pred = regressor.predict(X_test)

res = pd.DataFrame({"Actual": y_test,
                    "Predicted": y_pred,
                    "Difference": y_test - y_pred})
res
```

```python
from sklearn import metrics

print("MSE:", metrics.mean_squared_error(y_test, y_pred))
print("MAE:", metrics.mean_absolute_error(y_test, y_pred))
```

```python
metrics.mean_absolute_error(y_test, y_pred) / np.average(y) * 100
```

```python
regressor1 = DecisionTreeRegressor(criterion='mae')
regressor1.fit(X_train, y_train)

from sklearn import tree
tree.plot_tree(regressor1)
```

```python
y_pred1 = regressor1.predict(X_test)

res = pd.DataFrame({"Actual": y_test,
                    "Predicted": y_pred1,
                    "Difference": y_test - y_pred1})
res
```

```python
from sklearn import metrics
print("MSE:", metrics.mean_squared_error(y_test, y_pred1))
print("MAE:", metrics.mean_absolute_error(y_test, y_pred1))
```

```python
metrics.mean_absolute_error(y_test, y_pred1) / np.average(y) * 100
```
