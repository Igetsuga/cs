## Задание 2
Вывести 10 самых больших по размеру треков жанра Blues и формата MPEG (используйте команду LIMIT 10).


```python
%%sql
-- var. 1
SELECT 
    genres.GenreId as gId, media_types.MediaTypeId as mdTypeId, tracks.Bytes as size, tracks.Name
FROM 
    genres, media_types, tracks
WHERE 
    (tracks.genreId == gId) and (tracks.MediaTypeId == mdTypeId)
    and (gId == 6) and (mdTypeId == 1)
ORDER BY 
    tracks.Bytes DESC
LIMIT
    10;
```

```python
%%sql
-- var. 2
SELECT 
    tracks.Bytes as size, tracks.Name as name
FROM 
    tracks
WHERE 
    (tracks.genreId == 6) and (tracks.MediaTypeId == 1)
ORDER BY 
    tracks.Bytes DESC
LIMIT
    10;
```
